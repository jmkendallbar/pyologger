```python
import pandas as pd
import os

# Define the parameters
RecordingID = "Shuka_01-16-24"
Date = "20240116"  # Date in YYYYMMDD
starttime = "114015"  # Start time in HHMMSS
cameraNum = "96"
Nums = ["001", "002", "003"]  # List of file numbers

# Initialize an empty list to store DataFrames
data_frames = []

# Base path to your data directory, adjust as necessary
data_dir = "Data"

for Num in Nums:
    # Construct the filename for each file
    CATS_filename = f"{Date}-{starttime}-camera-{cameraNum}_{Num}.csv"
    # Construct the full file path
    file_path = os.path.join(data_dir, RecordingID, CATS_filename)
    
    # Read the CSV file into a DataFrame and append it to the list
    try:
        df = pd.read_csv(file_path, encoding='utf-8')  # Assuming UTF-8 encoding
        data_frames.append(df)
    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# Concatenate all DataFrames in the list into a single DataFrame
if data_frames:
    CATS_data = pd.concat(data_frames, ignore_index=True)
else:
    CATS_data = pd.DataFrame()  # Empty DataFrame if no files were read

# Now CATS_data contains all the data concatenated together
```
